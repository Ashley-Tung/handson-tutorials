```r
library(dplyr)
library(data.table)
library(scales)
library(ggrepel)
data = fread("20241231_TWAS_literature_review.csv")
data$publish_date <- as.Date(data$publish_date)
data <- data %>% arrange(publish_date)
data$type <- factor(data$type, levels = unique(data$type))
```

```r
# Set plot size for Jupyter
options(repr.plot.width = 30, repr.plot.height = 8)

# Create and display the plot again with dashed horizontal lines for each group
timeline_plot <- ggplot(data, aes(x = publish_date, y = type, label = method_name)) +
  # Add points for each method with matching color from the type
  geom_point(aes(color = type), size = 6) +  # Adjusted size for larger points
  
  # Increase text size and add text labels with better scatter control
  geom_text_repel(aes(label = method_name), size = 7, box.padding = 0.35, 
                  point.padding = 0.1, segment.color = 'black', 
                  min.segment.length = 0.1, max.overlaps = 20, show.legend = FALSE, vjust = -1.2) +
  
  # Set x-axis labels and breaks
  scale_x_date(labels = date_format("%Y-%m-%d"), breaks = "1 year") +
  
  labs(title = "Timeline of TWAS Methods", x = "Publish Date", y = "Type of method") +
  
  # Use minimal theme and adjust axis text, title, and grid lines
  theme_minimal() +
  ggplot2::theme(
    axis.text.x = element_text(size = 20, angle = 0, hjust = 0.5, face = "bold"),  # Bold x-axis
    axis.text.y = element_text(size = 15, face = "bold"),  # Bold y-axis
    axis.title.x = element_text(size = 20, face = "bold"),
    axis.title.y = element_text(size = 20, face = "bold"),
    axis.ticks.y = element_blank(),
    panel.grid.major.y = element_line(color = 'grey', linetype = 'dashed', linewidth = 0.5),  # Dashed horizontal lines
    panel.grid.minor.y = element_blank(),
    panel.grid.major.x = element_line(color = 'grey', linetype = 'dashed', linewidth = 0.5),  # Dashed vertical lines
    plot.title = element_text(hjust = 0.5, size = 30, face = "bold"),  # Bold title
    panel.grid.minor = element_blank(),
    
    # Adjust legend settings
    legend.text = element_text(size = 17),  # Increase legend text size
    legend.title = element_text(size = 20, face = "bold"),  # Bold legend title
    legend.key.size = unit(2, "cm"),  # Increase size of legend keys
    legend.position = "right"  # Optional: position legend to the right
  ) +
  
  # Add dashed horizontal lines for each type
  geom_hline(data = data, aes(yintercept = as.numeric(type)), 
             color = "black", linetype = "dashed", linewidth = 0.2) +
  
  # Add color scale to match the legend and points color, with reversed legend order
  scale_color_manual(values = c(
    "#D55C54",  # Red
    "#E67E3B",  # Orange
    "#FECB02",  # Yellow
    "#57C78B",  # Green
    "#71D6CA",  # Light blue
    "#4D9DE0",  # Blue
    "#7776BC",  # Purple
    "#FEC1C9"   # Light pink
  ),
  guide = guide_legend(reverse = TRUE))  # Reverse the legend order

timeline_plot
```

```r
# Save the plot to the specified directory
output_path <- "20241231_TWAS_literature_review_timeline_plot.png"
ggsave(output_path, plot = timeline_plot, width = 30, height = 8, dpi = 300)
```
